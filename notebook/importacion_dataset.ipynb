```python
import pandas as pd 
import numpy as np 

## Definimos la ruta donde se encuentra
ruta_jugador_info = r'C:\Users\Fernando\Desktop\CSV de NBA\jugador detalle.csv' 

## Leemos el dataset
df = pd.read_csv(ruta_jugador_info)

# Eliminar columnas innecesarias
columnas_a_eliminar = ['Unnamed: 0.4', 'Unnamed: 0.3', 'Unnamed: 0.2', 'Unnamed: 0.1', 'Unnamed: 0']
df = df.drop(columns=columnas_a_eliminar, errors='ignore')

def convertir_pies_a_cm(valor):
    try:
        # Separar en pies y pulgadas
        pies, pulgadas = str(valor).split("-")
        pies = int(pies)
        pulgadas = int(pulgadas)
        # Convertir a centÃ­metros
        altura_cm = pies * 30.48 + pulgadas * 2.54
        return round(altura_cm, 2)
    except:
        return None

df['Altura'] = df['Altura'].apply(convertir_pies_a_cm)

# FechaNacimiento: convertir a datetime
df['FechaNacimiento'] = pd.to_datetime(df['FechaNacimiento'], errors='coerce')

# JugoEstaTemporada: convertir a booleano (N = False, Y = True)
df['JugoEstaTemporada'] = df['JugoEstaTemporada'].map({'Y': True, 'N': False})


print(df)
```
