Importacion de dataset juego, para filtrar por aÃ±o.

```python
import pandas as pd 
import numpy as np 

ruta_archivo_csv = r'C:\Users\Fernando\Desktop\CSV de NBA\archive\csv\game.csv'
df_juego = pd.read_csv(ruta_archivo_csv)

print(f'Cantidad de equipos: {len(df_juego)}')
```

```python
# Cargamos nuevamente el data set, para "parsear" la columna 'game_date'.
df_juego = pd.read_csv(ruta_archivo_csv, parse_dates=['game_date'])

# Convertimos al formato fecha.
# df_games['game_date'] = pd.to_datetime(df_games['game_date'])
df_juego['game_date'] = pd.to_datetime(df_juego['game_date'], format='%Y-%m-%d')

print(df_juego['game_date'].dtype)


# Filtrar por fechas:
games_filtred = df_juego[
    (df_juego['game_date'] >= '2010-10-03') & ## 2010-10-03
    (df_juego['game_date'] <= '2020-10-06') ## 2020-10-06
]

# obtener el set de 'game_id' vÃ¡lidos
df_juego_id_validados = set(games_filtred['game_id'])
print(f'Cantidad de partidos entre 2010 y 2020: {len(df_juego_id_validados)}')

# # Muestro los tipos de datos:
# for columna in df_juego_id_validados.columns:
#     print(columna)
#     print(df_juego_id_validados[columna].describe())
#     print('------------------------------------')
```
